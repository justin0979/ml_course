```python
x = [1,2,3,4]
y = [11,22,88,100]
plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
ax.plot(x,y)
plt.show()
type(ax)
```

```python
# 2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([99999,10000,.5,.5])
ax.plot(x,y) # add some data
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, [50,100,200,250]); # add some data
type(fig), type(ax)
```

```python
type(fig), type(ax)
```

## Matplotlib example workflow

```python
import matplotlib.pyplot as plt
x = [100,2,3,45]
y = [20,18,34,44]

fig, ax = plt.subplots(figsize=(15,5))

ax.set(title="Name Me",
      xlabel="Now",
      ylabel='[OK]');

ax.plot(x,y);
```


## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np

x = np.linspace(0,10,100)
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),
      height=nut_butter_prices.values());

ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)")
plt.show()
```
