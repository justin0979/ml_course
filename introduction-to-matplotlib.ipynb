## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(15, 3)) # (width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Plot Title",
       xlabel="Bottom part",
       ylabel="Y side")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Subplots

#### Two options for subplots

### Subplot option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.set(xlabel="x", ylabel="x/2")
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

# Label each plot
ax1.set(title="Ax1")
ax2.set(title="Ax2")
ax3.set(title="Ax3", ylabel="Price ($)")
ax4.set(title="Ax4");
```

```python
# In above figure, titles are overlapping x-axis
#   - to fix, adjust figsize in plt.subplot()
#      - below, changed figsize=(10, 5) to figsize=(10, 7)
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 7))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));

# Label each plot
ax1.set(title="Ax1")
ax2.set(title="Ax2")
ax3.set(title="Ax3", ylabel="Price ($)")
ax4.set(title="Ax4");
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(15,5))
# Plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));

ax[0,0].set(title="ax[0,0]");
```

### Which plot method should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, ok to use the pyplot method
* When plotting somtheng more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
over_50.head()
```

### Pylot method

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10, 6));
```

### OO method mixed with Pyplot

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
# Increase the x-axis to start at 45 and end at 100 with ax.set_xlim([45, 100]),
#  currently, the xlim is at about [50, 80]
#  For this axes, 45,100 would create a lot of white space to the right
# ax.set_xlim([45, 100])
```

### OO method from scratch

```python
over_50["target"].values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

### Subplot of chol, age, thalach

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

# Customize ax0
ax0.set(title="Heart Disease: Age and Cholesterol Levels",
       # xlabel="Age"
        ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter_thalach = ax1.scatter(x=over_50["age"],
                              y=over_50["thalach"],
                              c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease: Age and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate (thalach)")
# Add a legend
ax1.legend(*scatter_thalach.legend_elements(), title="Target")
# add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

### Customizing the x-axis and y-axis limitations

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50.target,
                      cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Change the x-axis limits
# - This affects ax0 and ax1 because of `sharex=True` in plt.subplots()
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50.thalach,
                      c=over_50.target,
                      cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease and Heart Rate", ylabel="Heart Rate")

# Change ax1 x-axis limits
# - Below (`ax1.set_xlim()`) is not needed because of plt.subplots(sharex=True) from above
# ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Heart Rate")
# Add a mean line
ax1.axhline(y=over_50.thalach.mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
